## Study of the potential of different configurations

```python
# Number of configuration
n_config = len(C)**len(G)
# Number of states for each node
n_states = tuple(len(C) for _ in range(len(G)))

# Compute potential for each possible configuration
pot_per_conf = {}
for x_id in range(n_config):
    x = np.unravel_index(x_id,shape=n_states)
    conf = np.array(x)
    pot = potential(conf, W, cost)
    print("Configuration:")
    df = pd.DataFrame(data=[C_rev[z] for z in x], index=np.arange(1, len(G)+1), columns=['state']) # nodes start from 0
    df.index.name = 'node'
    display(df)
    print("POTENTIAL:", pot)
    print("\n")
    pot_per_conf[tuple(conf)] = pot # (convert 'conf' from numpy array to tuple since numpy array are not hashable)
```

```python
# Compute min and max of potential
min_pot = min(pot_per_conf.values())
min_conf = list(filter(lambda x: pot_per_conf[x]==min_pot, pot_per_conf))

printmd("Minimum potential $U$ = " + str(min_pot) + " - corresponding to configurations:")
for c in min_conf:
    print("Configuration:")
    df = pd.DataFrame(data=[C_rev[z] for z in c], index=np.arange(1, len(G)+1), columns=['state']) # nodes start from 0
    df.index.name = 'node'
    display(df)
    
max_pot = max(pot_per_conf.values())
max_conf = list(filter(lambda x: pot_per_conf[x]==max_pot, pot_per_conf))

printmd("Maximum potential $U$ = " + str(max_pot) + " - corresponding to configurations:")
for c in max_conf:
    print("Configuration:")
    df = pd.DataFrame(data=[C_rev[z] for z in c], index=np.arange(1, len(G)+1), columns=['state']) # nodes start from 0
    df.index.name = 'node'
    display(df)
```
